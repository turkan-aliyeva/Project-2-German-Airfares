```python
# Time to AM- and PM- format  
def convert_to_am_pm(time_str):
    if 'Uhr' in time_str:
        # Remove 'Uhr' from the string and strip whitespaces
        time_str = time_str.replace('Uhr', '').strip()

        # Convert the string to a datetime object
        time_obj = pd.to_datetime(time_str, format='%H:%M')

        # Convert the datetime object to the desired format
        return time_obj.strftime('%I:%M%p').lower()
    else:
        return time_str  # Return the original string if it doesn't contain "Uhr"

# Apply the revised conversion function to 'departure_time' and 'arrival_time' columns
df['departure_time_24h'] = df['departure_time'].apply(convert_to_am_pm)
df['arrival_time_24h'] = df['arrival_time'].apply(convert_to_am_pm)

# Display the rows that previously contained 'Uhr' to confirm the conversion
df[['departure_time', 'departure_time_24h', 'arrival_time', 'arrival_time_24h']].tail(20)
```

```python
# Time to 24 Hours Format 
def convert_to_24h_format(time_str):
    # Convert the string to a datetime object
    time_obj = pd.to_datetime(time_str, format='%I:%M%p')
    
    # Convert the datetime object to 24-hour format
    return time_obj.strftime('%H:%M')

# Convert the 'am/pm' columns to 24-hour format and store in new columns
df['departure_time_24h'] = df['departure_time_24h'].apply(convert_to_24h_format)
df['arrival_time_24h'] = df['arrival_time_24h'].apply(convert_to_24h_format)
df[['departure_time', 'departure_time_24h', 'arrival_time', 'arrival_time_24h']].head()
```

```python
df.head(20)
```

```python
# Convert the 'departure_time_24h' and 'arrival_time_24h' columns to datetime format
df['departure_time_24h'] = pd.to_datetime(df['departure_time_24h'], format='%H:%M')
df['arrival_time_24h'] = pd.to_datetime(df['arrival_time_24h'], format='%H:%M')

# Calculate the time difference
df['flight_duration (hr)'] = (df['arrival_time_24h'] - df['departure_time_24h']).dt.total_seconds() / 3600  # Convert seconds to hours

# Adjust cross midnight flights
df['flight_duration (hr)'] = df.apply(lambda row: row['flight_duration (hr)'] if row['departure_time_24h'] <= row['arrival_time_24h']
                                 else row['flight_duration (hr)'] + 24, axis=1)

# Convert 'departure_time_24h' and 'arrival_time_24h' back to string format without date
df['departure_time_24h'] = df['departure_time_24h'].dt.strftime('%H:%M')
df['arrival_time_24h'] = df['arrival_time_24h'].dt.strftime('%H:%M')

# Round the flight duration to two decimal places
df['flight_duration (hr)'] = df['flight_duration (hr)'].round(2)

df.tail()
```

```python
# Function to convert time intervals to days
def convert_to_days(interval):
    num, unit = interval.split()
    num = int(num)
    if unit == 'week' or unit == 'weeks':
        return num * 7
    elif unit == 'month'or unit == 'months':
        return num * 30
    elif unit == 'year' or unit == 'years':
        return num * 365
    else:
        return num

# Apply the function to create a new column with numeric representation
df['booking_days_before_departure'] = df['booking_time_before_departure'].apply(convert_to_days)

df
```

```python
# Function to convert stops to number
def convert_stops(stops):
    if stops == 'direct':
        return 0
    elif stops == '(1 stop)' or stops == '(1 Stopp)':
        return 1
    else:
        return 2

# Apply the function to create a new column with numeric representation
df['number_of_stops'] = df['stops'].apply(convert_stops)

df
```
